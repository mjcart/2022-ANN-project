```python
#작성자 전민재
#data augmentation
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#작성자 전민재
#데이터 load
X_train=np.uint8(np.load('./X_train.npy'))
y_train=np.uint8(np.load('./y_train.npy'))
X_valid=np.uint8(np.load('./X_valid.npy'))
y_valid=np.uint8(np.load('./y_valid.npy'))
#데이터 reshape
X_train=tf.reshape(X_train,[-1,28,28,1])
X_valid=tf.reshape(X_valid,[-1,28,28,1])
#데이터 resize
X_train_resize=np.uint8(tf.image.resize(X_train, [32, 32]))
X_valid_resize=np.uint8(tf.image.resize(X_valid, [32, 32]))
```

```python
#작성자 전민재
#최대 10도 회전 상하좌우 10%이동
def dataaugmentation(X,y):
    datagen=ImageDataGenerator(rotation_range=10,width_shift_range = 0.1, height_shift_range = 0.1)
    X_new=np.array([])
    y_new=np.array([])
    X_new=np.resize(X_new,[0,32,32,1])
    y_new=np.resize(y_new,[0])

    #영어 부분 2배 생성
    for i in range(2):
        #X_valid[:10]부분을 교체하고 실행
        for idx,image in enumerate(X):
            #숫자 제외하고 augmentation
            if y[idx]>9:

                #차원변환 
                #x=image.reshape(32,32,1)
                x=image[tf.newaxis,...]

                #새로운 데이터 생성
                image_result = next(iter(datagen.flow(x)))
                X_new=np.append(X_new,np.uint8((image_result)),axis=0)
                y_new=np.append(y_new,np.uint8(y[idx]),)
                
    return X_new,y_new
```

```python
def dataaugmentation(X,y):
    datagen=ImageDataGenerator(rotation_range=10,width_shift_range = 0.2, 
        height_shift_range = 0.2)

    X_new=[]
    y_new=[]
    #영어 부분 2배 생성
    for i in range(2):
        #X_valid[:10]부분을 교체하고 실행
        for idx,image in enumerate(X):
            #숫자 제외하고 augmentation
            if y[idx]>9:

                #차원변환 
                x=image.reshape(32,32,1)
                x=x[tf.newaxis,...]

                #새로운 데이터 생성
                image_result = next(iter(datagen.flow(x)))
                X_new.append(np.squeeze(image_result))
                y_new.append(y[idx])
    return X_new,y_new
```
