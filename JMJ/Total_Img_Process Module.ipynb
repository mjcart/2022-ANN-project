```python
#ImageResize Module
#작성자 전민재

#data 파일에 들어있는 여러이미지 일괄 변환해 resize파일에 저장


#초기 파일 위치에서 data파일로 이동
os.chdir('./data')
files=os.listdir()
#data 파일에 존재하는 원본 jpg나 png 파일에 접근 
for f in files:
    title, ext = os.path.splitext(f)
    if ext in ['.jpg', '.png']:
        img = Image.open(f)
        
         #lanczos필터를 사용해 높은 품질의 이미즈를 얻는다.
        img_resize = img.resize((380, 380),Image.LANCZOS)
        
        #resize디렉토리로 이동 후 저장
        os.chdir('./resize')
        print(os.getcwd() )
        img_resize.save( title+'resized' + ext)
        
        #다시 data 디렉토리로 복귀
        os.chdir("../")
#초기 디렉토리로 복귀
os.chdir("../")
```

```python
#ImageCuttingAndLabeling Module
#작성자 전민재

#imageresizemodule로 만들어진 파일들이 준비되어있어야한다.
#완성된 image cutting and 파일로 labeling
#모든 파일 문자별로 10개씩 쪼개고 파일 별 label

#현재 디렉토리 위치 출력
print(os.getcwd() )

#실행시 디렉토리 위치는 실행파일있는 곳으로 하자.


#resized된 파일이 있는곳으로 이동

os.chdir('./data/resize/')

#resized 파일
filelist=['num','AJ','aj_s','KT','kt_s','UZ','uz_s']

#label 파일 이름 리스트
AJ=[chr(_) for _ in range(65,75)]
aj_s=[f'{chr(_)}_s' for _ in range(97,107)]

KT=[chr(_) for _ in range(75,85)]
kt_s=[f'{chr(_)}_s' for _ in range(107,117)]


UZ=[chr(_) for _ in range(85,91)]
uz_s=[f'{chr(_)}_s' for _ in range(117,123)]
emptyspace=['empty','empty','empty','empty']
UZ=UZ+emptyspace
uz_s=uz_s+emptyspace

for afile in filelist:
    os.chdir(f'./{afile}')
    #디렉토리에 있는 파일 출력
    print(os.listdir())
    files=os.listdir()

    #파일에 접근 시작
    for idx, file in enumerate(files):
        fname, ext = os.path.splitext(file)

        #디렉토리에있는 파일중 jpg,png파일만 읽어들임
        if ext in ['.jpg','.png']:
            path=os.getcwd()
            print(path,file)
            im = Image.open(file)
            os.chdir('../') # resize dir 로 이동임
            
            for i in range(10):
                #해당 Label파일에 cutting된 사진 저장 
                if afile =='num':
                    os.chdir(f"./{i}")
                elif afile=='AJ':
                    os.chdir(f"./{AJ[i]}")
                elif afile=='aj_s':
                    os.chdir(f"./{aj_s[i]}")
                elif afile=='KT':
                    os.chdir(f"./{KT[i]}")
                elif afile=='kt_s':
                    os.chdir(f"./{kt_s[i]}")
                elif afile=='UZ':
                    os.chdir(f"./{UZ[i]}")
                elif afile=='uz_s':
                    os.chdir(f"./{uz_s[i]}")
                    
                for j in range(10):
                    #(x,y,x+width,y+height)
                    crop_image = im.crop((20+i*34, 20+j*34, 54+i*34 , 54+j*34))

                    #pixel값에 접근하기 위해 load (crop_image와 연결되어 있어 값을 바꾸기만하면됨)
                    load_image=crop_image.load()

                    #테두리 침식
                    for x in [0,1,2,31,32,33]:
                        for y in range(34):
                            load_image[x,y]=(255,255,255)
                    for y in [0,1,2,31,32,33]:
                         for x in range(34):
                            load_image[x,y]=(255,255,255)   

                    #28*28resize
                    crop_image = crop_image.resize((28, 28),Image.LANCZOS)

                    #gray 변환
                    crop_image=crop_image.convert('L')

                    #여기 이름 새로추가 5/11
                    if afile =='num':
                        imgname=f"{i}"
                    elif afile=='AJ':
                        imgname=f"{AJ[i]}"
                    elif afile=='aj_s':
                        imgname=f"{aj_s[i]}"
                    elif afile=='KT':
                        imgname=f"{KT[i]}"
                    elif afile=='kt_s':
                        imgname=f"{kt_s[i]}"
                    elif afile=='UZ':
                        imgname=f"{UZ[i]}"
                    elif afile=='uz_s':
                        imgname=f"{uz_s[i]}"
                    crop_image.save(imgname + '_data_'+str(idx*10+j) + '.jpg')
                    print(imgname + '_data_' +str(j) + '.jpg저장')

                #모두 잘라낸뒤 resize파일로 돌아가기
                os.chdir("../")
            os.chdir(f"./{afile}")
    os.chdir("../")
    # 실행이 끝나면 resize파일이다.
    #다시 기본디렉토리로 이동
os.chdir("../")
os.chdir("../")

```
